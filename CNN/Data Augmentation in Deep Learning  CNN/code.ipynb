```python
from keras.preprocessing import image
from keras.src.legacy.preprocessing.image import ImageDataGenerator
```

```python
img = image.load_img('./image/download.jpeg', target_size=(200,200))
```

```python
img
```

```python
type(img)
```

```python
datagen = ImageDataGenerator(
        rotation_range=30,
        width_shift_range=0.2,
        height_shift_range=0.2,
        shear_range=0.2,
        zoom_range=0.2,
        horizontal_flip=True,
        fill_mode='nearest')
```

```python
img = image.img_to_array(img)
```

```python
type(img)
```

```python
img.shape
```

```python
input_batch = img.reshape(1, 200,200,3)
```

```python
i = 0

for outout in datagen.flow(input_batch, batch_size=1,save_to_dir='image'):
    
    i = i + 1
    
    if i == 10:
        break
```
